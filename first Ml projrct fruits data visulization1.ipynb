```python
from os import X_OK
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np


# lets add dataset of fruits

data = {
    'Fruits' :['Apple','Banana','Cherries','Dates'],
    'Quantity' :[25,10,15,30],
    'Price' :[11,10,3,2],
}
# now raw data into pandas dataframes
df = pd.DataFrame(data)
# print(df)
# multiply qunatity and price into new column
df['Total_Price'] = df['Quantity'] * df['Price']

print(df)

# Now print all values
print("\n-------Lets Draw the chat--------")
print("\n")
# going to plot this data in Bar chart
plt.bar(df['Fruits'], df['Total_Price'])
plt.title("Total sales per Fruit")
plt.xlabel("Fruits")
plt.xlabel("Total_Price")

plt.show()

print("\n")
# now will check in line chart
plt.plot(df['Fruits'], df['Total_Price'], marker='o')
plt.title("Sales Trend")
plt.xlabel("Fruits")
plt.ylabel("Total Price")
plt.show()

print("\n")
# now lets try in pie chart
plt.pie(df['Total_Price'], labels=df['Fruits'], autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()
print("\n")

# simple ml model for
# Predict Total price from Qunatity
print("\n")

# X = df[['Quantity']]
# y = df[['Total_Price']]
# model = LinearRegression
# model.fit(X,y)

# lets predict for new value
# new_quantity =np.array([[40]])
# prediction = model.predict(new_quantity)
# print("\n Predicted Toatal price for Quantity=40:", prediction[0])




```
